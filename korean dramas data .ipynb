```python
import matplotlib.pyplot as plt

# Sample data
Drama_Name = ['All of us are death', 'Legend of the Blue Sea', 'The Heirs', 'Boys over Flowers']
ViewshipRate = [5, 17.6, 16.7, 25.7]

# Sorting data
Drama_Name.reverse()
ViewshipRate.sort(reverse=True)

# Create x coordinates using indices of labels
x = range(len(Drama_Name))

# Plot the funnel chart
plt.figure(figsize=(8, 6))
plt.barh(x, ViewshipRate, color=["skyblue", "purple", "black", "green"], alpha=0.6)

# Add labels and title
plt.yticks(x, Drama_Name)  # Set y-axis labels to Korean drama names
plt.xlabel('Viewship Rate')
plt.ylabel('Drama Name')
plt.title('Korean Drama Funnel')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
Drama_Name = ['All of us are death', 'Legend of the Blue Sea', 'The Heirs', 'Boys over Flowers']
Network = [5, 17.6, 16.7, 25.7]

# Plot the bar chart
plt.figure(figsize=(8, 6))
plt.bar(Drama_Name, Network, color='green')

# Add labels and title
plt.xlabel('Drama Name')
plt.ylabel('Network')
plt.title('Korean Drama Network Distribution')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
Drama_Name = ['Extraordinary You', 'Boys over Flowers', 'A love so beautiful', 'Itâ€™s okay to not be okay']
No_of_Episodes = [32, 25, 24, 16]

# Plot the pie chart
plt.figure(figsize=(8, 6))
plt.pie(No_of_Episodes, labels=Drama_Name, autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'lightcoral', 'lightskyblue'])

# Draw a white circle at the center to make it a donut chart
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# Add a title
plt.title('Korean Drama No of Episodes Distribution')

# Equal aspect ratio ensures that pie is drawn as a circle
plt.axis('equal')

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
Drama_Name = ['All of us are death', 'Legend of the Blue Sea', 'The Heirs', 'Boys over Flowers']
Network = [5, 17.6, 16.7, 25.7]

# Plot the exponential graph
plt.figure(figsize=(8, 6))
plt.plot(Drama_Name, np.exp(Network), color="green")

# Add labels and title
plt.xlabel('Drama Name')
plt.ylabel('Network (Exponential)')
plt.title('Exponential Plot of Korean Drama Network Distribution')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.grid(True)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data
Drama_Name = ['All of us are death', 'Legend of the Blue Sea', 'The Heirs', 'Boys over Flowers']
Network = [5, 17.6, 16.7, 25.7]

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.hist(Network, bins=10, color="purple", edgecolor='red', alpha=0.7)

# Add labels and title
plt.xlabel('Network')
plt.ylabel('Frequency')
plt.title('Histogram of Korean Drama Network Distribution')

# Show the plot
plt.grid(True)
plt.show()
```
